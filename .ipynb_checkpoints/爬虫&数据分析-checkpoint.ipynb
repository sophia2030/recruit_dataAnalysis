```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import re
import os
import glob
from collections import Counter
from datetime import datetime

# ============================================================
# ===== 配置参数（修改这里即可）=====
# ============================================================
DATA_DIR = "./data"           # 数据目录，留空则使用脚本所在目录
RECURSIVE_SCAN = False  # 是否扫描子目录
# ============================================================
# ===== 以下为程序主体（无需修改）=====
# ============================================================

plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 停用词列表（可自动扩展）
STOP_WORDS = {
    '负责', '能够', '具备', '具有', '优先', '相关', '工作', '职位', '岗位', 
    '熟悉', '掌握', '了解', '良好的', '较强的', '一定的', '优秀的',
    '经验', '学历', '本科', '硕士', '不限', '及以上', '优先考虑',
    '我们', '提供', '要求', '职责', '任职资格', '岗位职责'
}

def scan_json_files(directory, recursive=False):
    """扫描目录下所有JSON文件"""
    pattern = os.path.join(directory, "**/*.json" if recursive else "*.json")
    files = glob.glob(pattern, recursive=recursive)
    exclude_patterns = ["eda_report", "_progress", "cleaned"]
    return [f for f in files if not any(p in os.path.basename(f) for p in exclude_patterns)]

def extract_salary_mid(s):
    """提取薪资中值"""
    if pd.isna(s):
        return None
    nums = re.findall(r'(\d+\.?\d*)', str(s))
    if len(nums) >= 2:
        return (float(nums[0]) + float(nums[1])) / 2
    elif len(nums) == 1:
        return float(nums[0])
    return None

def extract_exp_info(s):
    """提取经验年限"""
    if pd.isna(s):
        return None, None
    nums = re.findall(r'(\d+)', str(s))
    if len(nums) >= 2:
        return int(nums[0]), int(nums[1])
    elif len(nums) == 1:
        return int(nums[0]), int(nums[0])
    return None, None

def extract_skills_smart(texts, min_freq=3, top_n=30):
    """
    智能提取技能关键词
    基于词频 + 词性过滤 + 共现分析
    """
    # 简单分词：按中文和英文词汇拆分
    all_words = []
    for text in texts:
        if pd.isna(text):
            continue
        # 英文单词
        eng_words = re.findall(r'[A-Za-z][A-Za-z0-9\+\#\.]+[A-Za-z]', text)
        # 中文词组（2-4个字）
        cn_words = re.findall(r'[\u4e00-\u9fa5]{2,4}', text)
        all_words.extend([w.lower() for w in eng_words])
        all_words.extend(cn_words)
    
    # 过滤停用词
    filtered = [w for w in all_words if w not in STOP_WORDS and len(w) > 1]
    
    # 统计词频
    counter = Counter(filtered)
    
    # 过滤低频词并排序
    result = {word: count for word, count in counter.items() 
              if count >= min_freq}
    sorted_result = sorted(result.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_result[:top_n]

def load_words_from_dir(directory, default_list=None, file_patterns=["*.txt", "*.json"]):
    """
    从指定目录加载词汇，合并所有 .txt 和 .json 文件中的词条。
    如果目录不存在或没有匹配文件，返回 default_list（默认为空列表）。
    """
    if default_list is None:
        default_list = []
    words = set(default_list)
    
    if not os.path.exists(directory):
        print(f"📂 目录不存在: {directory}，使用默认词库（{len(words)} 个）")
        return words

    found_files = False
    for pattern in file_patterns:
        for filepath in glob.glob(os.path.join(directory, pattern)):
            found_files = True
            try:
                if filepath.endswith('.json'):
                    with open(filepath, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                    # 提取字符串元素
                    if isinstance(data, list):
                        for item in data:
                            if isinstance(item, str):
                                words.add(item.strip())
                            elif isinstance(item, dict) and 'name' in item:
                                words.add(item['name'].strip())
                    elif isinstance(data, dict):
                        # 递归提取所有 name 字段
                        def extract(obj):
                            if isinstance(obj, dict):
                                if 'name' in obj:
                                    words.add(obj['name'].strip())
                                for v in obj.values():
                                    extract(v)
                            elif isinstance(obj, list):
                                for item in obj:
                                    extract(item)
                        extract(data)
                else:
                    # txt 文件：每行一个词
                    with open(filepath, 'r', encoding='utf-8') as f:
                        for line in f:
                            word = line.strip()
                            if word:
                                words.add(word)
                print(f"✅ 已加载: {os.path.basename(filepath)}")
            except Exception as e:
                print(f"⚠️ 读取文件 {filepath} 失败: {e}")

    if not found_files:
        print(f"📂 目录 {directory} 下无匹配文件，使用默认词库（{len(words)} 个）")
    else:
        print(f"📊 共加载 {len(words)} 个词汇（来自 {directory}）")
    return words

def analyze_single_file(filepath):
    """分析单个JSON文件"""
    print(f"\n{'='*60}")
    print(f"📁 分析: {os.path.basename(filepath)}")
    print(f"{'='*60}")
    
    with open(filepath, 'r', encoding='utf-8') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    
    if df.empty:
        print("⚠️ 数据为空")
        return None
    
    print(f"📊 总记录: {len(df)} | 字段: {list(df.columns)}")
    
    # 薪资
    if 'salary' in df.columns:
        df['salary_mid'] = df['salary'].apply(extract_salary_mid)
        valid_salary = df['salary_mid'].dropna()
        if not valid_salary.empty:
            print(f"💰 薪资中位数: {valid_salary.median():.1f}K | 范围: {valid_salary.min():.1f}-{valid_salary.max():.1f}K")
    
    # 经验
    if 'experience' in df.columns:
        df['exp_min'], df['exp_max'] = zip(*df['experience'].apply(extract_exp_info))
        valid_exp = df['exp_min'].dropna()
        if not valid_exp.empty:
            print(f"📅 平均经验要求: {valid_exp.mean():.1f}年")
    
    # 🧠 智能技能提取
    if 'jobDesc' in df.columns:
        skills = extract_skills_smart(df['jobDesc'].tolist(), min_freq=2, top_n=20)
        if skills:
            print(f"\n🧠 智能识别的技能关键词 TOP 50:")
            for word, count in skills[:50]:
                pct = count / len(df) * 100
                print(f"  {word}: {count}次 ({pct:.1f}%)")
    
    # 城市分布
    if 'city' in df.columns:
        cities = df['city'].value_counts().head(3)
        if not cities.empty:
            print(f"\n📍 主要城市: {', '.join([f'{c}({n}条)' for c,n in cities.items()])}")
    
    # 学历
    if 'education' in df.columns:
        edu_counts = df['education'].value_counts().head(3)
        if not edu_counts.empty:
            print(f"\n🎓 学历要求: {', '.join([f'{e}({n}条)' for e,n in edu_counts.items()])}")
    
    return df

#加载词库
DEFAULT_STOPWORDS = []   # 或者放一些基础词，如 ['负责', '能够', ...]
STOP_WORDS = load_words_from_dir("wordlists/stopwords", default_list=DEFAULT_STOPWORDS)

# 你可以在这里放一些基础技能词（留空也行）
DEFAULT_SKILLS = []      # 或者放一些基础技能，如 ['python', 'sql', ...]
SKILL_KEYWORDS = load_words_from_dir("wordlists/skills", default_list=DEFAULT_SKILLS)

def main():
    data_dir = DATA_DIR if DATA_DIR else os.getcwd()
    print(f"🔍 扫描: {data_dir}")
    
    json_files = scan_json_files(data_dir, RECURSIVE_SCAN)
    
    if not json_files:
        print("⚠️ 未找到JSON文件")
        return
    
    print(f"📂 找到 {len(json_files)} 个文件")
    all_dfs = []
    
    for f in json_files:
        df = analyze_single_file(f)
        if df is not None:
            all_dfs.append(df)
            # 自动保存清洗后的CSV
            csv_path = f"{os.path.splitext(f)[0]}_cleaned.csv"
            df.to_csv(csv_path, index=False, encoding='utf-8-sig')
            print(f"💾 已保存: {csv_path}")
    
    if all_dfs:
        combined = pd.concat(all_dfs, ignore_index=True)
        print(f"\n{'='*60}")
        print(f"📊 总览: 共分析 {len(all_dfs)} 个文件，{len(combined)} 条记录")
    
    print("\n✅ 完成!")

if __name__ == "__main__":
    main()
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import re
import os
import glob
from collections import Counter
from datetime import datetime

# ============================================================
# ===== 配置参数（修改这里即可）=====
# ============================================================
DATA_DIR = "./data"           # 数据目录，留空则使用脚本所在目录
RECURSIVE_SCAN = False  # 是否扫描子目录
# ============================================================
# ===== 以下为程序主体（无需修改）=====
# ============================================================

plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 停用词列表（可自动扩展）
STOP_WORDS = {
    '负责', '能够', '具备', '具有', '优先', '相关', '工作', '职位', '岗位', 
    '熟悉', '掌握', '了解', '良好的', '较强的', '一定的', '优秀的',
    '经验', '学历', '本科', '硕士', '不限', '及以上', '优先考虑',
    '我们', '提供', '要求', '职责', '任职资格', '岗位职责'
}

def scan_json_files(directory, recursive=False):
    """扫描目录下所有JSON文件"""
    pattern = os.path.join(directory, "**/*.json" if recursive else "*.json")
    files = glob.glob(pattern, recursive=recursive)
    exclude_patterns = ["eda_report", "_progress", "cleaned"]
    return [f for f in files if not any(p in os.path.basename(f) for p in exclude_patterns)]

def extract_salary_mid(s):
    """提取薪资中值"""
    if pd.isna(s):
        return None
    nums = re.findall(r'(\d+\.?\d*)', str(s))
    if len(nums) >= 2:
        return (float(nums[0]) + float(nums[1])) / 2
    elif len(nums) == 1:
        return float(nums[0])
    return None

def extract_exp_info(s):
    """提取经验年限"""
    if pd.isna(s):
        return None, None
    nums = re.findall(r'(\d+)', str(s))
    if len(nums) >= 2:
        return int(nums[0]), int(nums[1])
    elif len(nums) == 1:
        return int(nums[0]), int(nums[0])
    return None, None

def extract_skills_with_filter(
    texts,
    skill_set=None,
    stopwords=None,
    min_freq=3,
    top_n=30,
    strict_whitelist=False,
    calc_proficiency=True   # 新增：设为False可彻底关闭熟练度检测，恢复旧版速度
):
    """从文本列表提取技能，应用黑白名单过滤，并计算熟练度（支持性能开关）"""
    if skill_set is None:
        skill_set = set()
    if stopwords is None:
        stopwords = set()

    prof_weight = {
        '精通': 5, '熟练掌握': 4, '熟练': 3,
        '熟悉': 2, '了解': 1, '掌握': 3
    }

    skill_counter = Counter()
    prof_map = {}

    for text in texts:
        if pd.isna(text):
            continue

        # 提取候选词
        eng_words = re.findall(r'[A-Za-z][A-Za-z0-9\+\#\.\-]+[A-Za-z]', text)
        cn_words = re.findall(r'[\u4e00-\u9fa5]{2,4}', text)
        candidates = set(w.lower() for w in eng_words + cn_words)

        # 黑白名单过滤
        final_words = set()
        for word in candidates:
            if word in skill_set:
                final_words.add(word)
            elif word in stopwords:
                continue
            else:
                if not strict_whitelist:
                    final_words.add(word)

        # 统计 + 熟练度检测（优化版）
        for word in final_words:
            # ⚡ 核心优化：先用字符串快速检查，避免无效正则扫描
            if calc_proficiency:
                if word not in text.lower():
                    # 理论上 final_words 来自 text，但为了防止大小写/边界误差，兜底
                    prof_map.setdefault(word, []).append(0)
                    skill_counter[word] += 1
                    continue
                # 确认识别到词，再提取前面的修饰词
                pattern = rf'(.{{0,10}})\b{re.escape(word)}\b'
                match = re.search(pattern, text, re.IGNORECASE)
                if match:
                    prefix = match.group(1).strip()
                    weight = 0
                    for k, v in prof_weight.items():
                        if k in prefix:
                            weight = v
                            break
                    prof_map.setdefault(word, []).append(weight)
                else:
                    prof_map.setdefault(word, []).append(0)
            else:
                # 不计算熟练度，仅计数（极速模式）
                prof_map.setdefault(word, []).append(0)
            
            skill_counter[word] += 1

    # 生成结果
    result = []
    for word, cnt in skill_counter.items():
        if cnt < min_freq:
            continue
        weights = prof_map.get(word, [0]*cnt)
        avg_prof = sum(weights) / len(weights) if weights else 0
        importance = cnt * (1 + avg_prof / 5) if calc_proficiency else cnt
        result.append({
            'skill': word,
            'frequency': cnt,
            'avg_proficiency': round(avg_prof, 2),
            'importance': round(importance, 2)
        })

    result.sort(key=lambda x: x['importance'], reverse=True)
    return result[:top_n]

def load_words_from_dir(directory, default_list=None, file_patterns=["*.txt", "*.json"]):
    """
    从指定目录加载词汇，合并所有 .txt 和 .json 文件中的词条。
    如果目录不存在或没有匹配文件，返回 default_list（默认为空列表）。
    """
    if default_list is None:
        default_list = []
    words = set(default_list)
    
    if not os.path.exists(directory):
        print(f"📂 目录不存在: {directory}，使用默认词库（{len(words)} 个）")
        return words

    found_files = False
    for pattern in file_patterns:
        for filepath in glob.glob(os.path.join(directory, pattern)):
            found_files = True
            try:
                if filepath.endswith('.json'):
                    with open(filepath, 'r', encoding='utf-8') as f:
                        data = json.load(f)
                    # 提取字符串元素
                    if isinstance(data, list):
                        for item in data:
                            if isinstance(item, str):
                                words.add(item.strip())
                            elif isinstance(item, dict) and 'name' in item:
                                words.add(item['name'].strip())
                    elif isinstance(data, dict):
                        # 递归提取所有 name 字段
                        def extract(obj):
                            if isinstance(obj, dict):
                                if 'name' in obj:
                                    words.add(obj['name'].strip())
                                for v in obj.values():
                                    extract(v)
                            elif isinstance(obj, list):
                                for item in obj:
                                    extract(item)
                        extract(data)
                else:
                    # txt 文件：每行一个词
                    with open(filepath, 'r', encoding='utf-8') as f:
                        for line in f:
                            word = line.strip()
                            if word:
                                words.add(word)
                print(f"✅ 已加载: {os.path.basename(filepath)}")
            except Exception as e:
                print(f"⚠️ 读取文件 {filepath} 失败: {e}")

    if not found_files:
        print(f"📂 目录 {directory} 下无匹配文件，使用默认词库（{len(words)} 个）")
    else:
        print(f"📊 共加载 {len(words)} 个词汇（来自 {directory}）")
    return words

def analyze_single_file(filepath):
    """分析单个JSON文件"""
    print(f"\n{'='*60}")
    print(f"📁 分析: {os.path.basename(filepath)}")
    print(f"{'='*60}")
    
    with open(filepath, 'r', encoding='utf-8') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    
    if df.empty:
        print("⚠️ 数据为空")
        return None
    
    print(f"📊 总记录: {len(df)} | 字段: {list(df.columns)}")
    
    # 薪资
    if 'salary' in df.columns:
        df['salary_mid'] = df['salary'].apply(extract_salary_mid)
        valid_salary = df['salary_mid'].dropna()
        if not valid_salary.empty:
            print(f"💰 薪资中位数: {valid_salary.median():.1f}K | 范围: {valid_salary.min():.1f}-{valid_salary.max():.1f}K")
    
    # 经验
    if 'experience' in df.columns:
        df['exp_min'], df['exp_max'] = zip(*df['experience'].apply(extract_exp_info))
        valid_exp = df['exp_min'].dropna()
        if not valid_exp.empty:
            print(f"📅 平均经验要求: {valid_exp.mean():.1f}年")
    
    # 🧠 技能提取（使用黑白名单过滤）
    if 'jobDesc' in df.columns:
        skills = extract_skills_with_filter(
            texts=df['jobDesc'].tolist(),
            skill_set=SKILL_KEYWORDS,
            stopwords=STOP_WORDS,
            min_freq=2,
            top_n=20,
            strict_whitelist=False,
            calc_proficiency=True   # 想要熟练度分析就保持 True；觉得慢就改成 False
        )
        if skills:
            print(f"\n🧠 技能与熟练度分析 (按重要程度排序):")
            print(f"{'技能':<15} {'出现次数':<10} {'平均熟练度':<12} {'重要程度'}")
            print("-" * 50)
            for item in skills:
                print(f"{item['skill']:<15} {item['frequency']:<10} {item['avg_proficiency']:<12} {item['importance']}")
    
    # 城市分布
    if 'city' in df.columns:
        cities = df['city'].value_counts().head(3)
        if not cities.empty:
            print(f"\n📍 主要城市: {', '.join([f'{c}({n}条)' for c,n in cities.items()])}")
    
    # 学历
    if 'education' in df.columns:
        edu_counts = df['education'].value_counts().head(3)
        if not edu_counts.empty:
            print(f"\n🎓 学历要求: {', '.join([f'{e}({n}条)' for e,n in edu_counts.items()])}")
    
    return df

#加载词库
DEFAULT_STOPWORDS = []   # 或者放一些基础词，如 ['负责', '能够', ...]
STOP_WORDS = load_words_from_dir("wordlists/stopwords", default_list=DEFAULT_STOPWORDS)

# 你可以在这里放一些基础技能词（留空也行）
DEFAULT_SKILLS = []      # 或者放一些基础技能，如 ['python', 'sql', ...]
SKILL_KEYWORDS = load_words_from_dir("wordlists/skills", default_list=DEFAULT_SKILLS)

def main():
    data_dir = DATA_DIR if DATA_DIR else os.getcwd()
    print(f"🔍 扫描: {data_dir}")
    
    json_files = scan_json_files(data_dir, RECURSIVE_SCAN)
    
    if not json_files:
        print("⚠️ 未找到JSON文件")
        return
    
    print(f"📂 找到 {len(json_files)} 个文件")
    all_dfs = []
    
    for f in json_files:
        df = analyze_single_file(f)
        if df is not None:
            all_dfs.append(df)
            # 自动保存清洗后的CSV
            csv_path = f"{os.path.splitext(f)[0]}_cleaned.csv"
            df.to_csv(csv_path, index=False, encoding='utf-8-sig')
            print(f"💾 已保存: {csv_path}")
    
    if all_dfs:
        combined = pd.concat(all_dfs, ignore_index=True)
        print(f"\n{'='*60}")
        print(f"📊 总览: 共分析 {len(all_dfs)} 个文件，{len(combined)} 条记录")
    
    print("\n✅ 完成!")

if __name__ == "__main__":
    main()
```
